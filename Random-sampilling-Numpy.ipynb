# random.randint()

this function use to give the integer value 
means give grater then 0 value 

```python
# for integer number mens greater then 1

np.random.randint(1,4) # this function use for normal random numbers 1 D 
```

```python
# if we want random matrix grayer then zero then we use this function

np.random.randint(1,4,(4,4))

# Here the random matrix generated  2 D Array 
```

```python
np.random.randint(1,4,(2,4,4))

# 3 D Array is generated with the help of this function 
```

# random.seed()

seed function give the same out put same as previous function 

```python
# when we use the random function the value are continuously change in every out put

# if we want same out put same as previous function then we use this type of function

np.random.seed(10)
np.random.randint(1,4,(2,4,4))
```

```python
np.random.seed(10)
np.random.randint(1,4,(2,4,4))
```

```python
# both the 3 D array is same 
```
