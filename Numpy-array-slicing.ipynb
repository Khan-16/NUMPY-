# Numpy Array Slicing (:)

```python
# array slicing use to find out the particular terms from the array 
```

```python
import numpy as np 
```

```python
# here we make 2 D array with 10 rows and 10 coloumn

mx = np.arange(1,101).reshape(10,10)
print(mx)
```

```python
mx[0,0] # Here the slicing is 1

# we have to check first number of rows and number of column 
# we have to find 1 then here 1 is on zeroth row and zeroth column 
```

```python
mx[4,4] #for checking the we use the square bracket []
        # for particular number access 
```

```python
mx[4,4].ndim # here we use .ndim to check the diamension 
```

```python
mx[0] #here the print is single row 
      # for particular row acces 
```

```python
mx[1]
```

```python
mx[:,0] #  here we use the slice operator :,
         # 1 D column we get 
        # for particular column access 
```

```python
 # we find two D

mx[:,0:1] # particular column acces with 2 D
```

```python
mx[:,0:1].ndim # checking the diamension of particular column 
```

```python
mx
```

```python
# if we want random rows and random colum the syntex is 

mx[1:4,1:4]
```

```python
mx[:,1:3] #here the total rows and only two columns 
```

```python
mx[:] # full matrix acces with this tecnique
      # matrix access with single slice operator 

```

```python
mx[::] # second method to acces the column 
      # matrix access with double slice operator 
```

```python
mx[:,:] # third method to acces the column 
```

```python
# for finding the particular data size

mx.itemsize # data size of particular matrix 
```
