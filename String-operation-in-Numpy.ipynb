# String Operations,Comparison and information in Numpy 

```python
import numpy as np 
```

# Add two String 

```python
# now we generate two random sting for string operation 

ch_name1 = "mohd yusuf raza khan"

ch_name2 = "is a good boy"
```

```python
np.char.add(ch_name1,ch_name2)
```

# lower case 

```python
# string convert lower case 

np.char.lower(ch_name1)
```

# uppar case 

```python
# string convert upper case

np.char.upper(ch_name1)
```

# middle 

```python
# convert string in to center

np.char.center(ch_name1,60)

# now the name is in cenetr 
```

```python
# space fill 

np.char.center(ch_name1,60,fillchar="-")
```

# split string 


```python
np.char.split(ch_name1)
```

```python
np.char.splitlines("hello\nkhan") # both values are splited 
```

# join string 

```python
# yek string ko dusre string se join karna hai to ye join function ka use karenge 
# join string with specific character 

str4 = "dmy"
str5 = "dmy"
```

```python
np.char.join([":","/"],[str4,str5]) 
```

```python
# join function use in name

ch = "mohd yusuf raza khan"
ch1 = "i am from mumbai"
```

```python
np.char.join('', [ch, ch1]) # yek particular character ko particular character se join 
```

# REPLACE 

```python
# replace the particular string to an other string 

np.char.replace(ch1,"mumbai","Lucknow")
```

# comparision

```python
# check the particular string is equalor not 

np.char.equal(ch,ch1)

# this is showing not equaal 
```
