# Mathematic Operation Using Numpy

```python
import numpy as np 
```

```python
arr1 = np.arange(1,10).reshape(3,3)
arr2 = np.arange(1,10).reshape(3,3)
print(arr1)
print(arr2)
```

# Addition 


```python
# Addition of two matrix 

arr1 + arr2 
```

```python
# we also yse  a function to add two matrix 

np.add(arr1,arr2 )
```

# Substraction 


```python
# substract one matrix to an other matrix

arr1-arr2 # normal formate 
```

```python
# functions for substraction

np.subtract(arr1,arr2)
```

# Divide

```python
# Divide one matrix to one other matrix 
arr1/arr2
```

```python
# divide function
np.divide(arr1,arr2)
```

# Multiplication

```python
# multiplication from one elemnet to an other element

arr1*arr2
```

```python
# multiplication function

np.multiply(arr1,arr2) # here are the simplemultiplication from one element to an other element 

# for matrix multiplication we have to use @ function
```

```python
arr1@arr2 # this is the real matrix function multiplication 
```

```python
# function

arr1.dot(arr2)
```
