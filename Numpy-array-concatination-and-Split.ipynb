# Numpy Array Concatination and Split 

```python
# is function me hum deakhenge ki hum array ko kaise join karte hai aur particular array ko kaise split karte hai 
```

```python
import numpy as np 
```

Humne yaha yek rendom data ko generate kiya hai 

```python
ar_1 = np.arange(1,17).reshape(4,4) #humne isme reshape condition laga ke matrix generate kiya hai 
print(ar_1)
```

```python
ar_2 = np.arange(17,33).reshape(4,4) #humne isme reshape condition laga ke matrix generate kiya hai 
print(ar_2)
```

# Concatenate

first array ko hum second array ke sath hum conctinate karenge 

Means first array ke just nechey second array ayega 

Humko yaha par add nahi karna hai yaha par dono arrays ko jaoin karna hai 

# column wise concatenation 

```python
# yaha par donoarray ko hum concatenate karenge with the help of this function

np.concatenate((ar_1,ar_2)) #  concatenate karne ke liye humko do braket lagana padta hai 

# ye humne coloumn wise concatenation kiya hai 
# means first array ka columnn dusre array ke column me add ho gaya hai
```

# Row wise concatenation 

```python
print(ar_1)

```

```python
print(ar_2)
```

```python
# yaha par donoarray ko hum concatenate karenge with the help of this function

np.concatenate((ar_1,ar_2),axis = 1) #  concatenate karne ke liye humko do braket lagana padta hai and uske sath axis bhi define karna padta hai for rows concatenation  

# ye humne Row wise concatenation kiya hai 
# means first array ka Row dusre array ke Row me add ho gaya hai
```

```python
# A Shortcut method of column wise concatenation 

np.vstack((ar_1,ar_2)) #yaha par bhi humko tupple ka use karna padega mens double bracket
```

```python
# A Shortcut method of Row wise concatenation 

np.hstack((ar_1,ar_2))
```

# Split

```python
# array ko split karne ka functio

np.split(ar_1,2) # comma 2 means ki humko particular array kitne part me divide karna hai 
```
