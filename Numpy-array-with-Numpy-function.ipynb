# Reshaped 

reshape use to re design the 1 D array in multi D array 

```python
# made 1 D array 
import numpy as np
ar_1d = np.arange(1,13)
print(ar_1d) 
print(ar_1d.ndim) # this is one diamensional array and now converted in two multi diamensional 
```

```python
# we use the reshaped functionn 
# 1 D array converted in to 2 D 

ar_2d = ar_1d.reshape(2,4)
print(ar_2d) # this converts in two two diamensional array 
```

```python
#  1D array converted in to 3D

ar_3d = ar_1d.reshape(2,3,2)
print(ar_3d)
```

```python
# a single line array function converted in to 2 D

ar = np.arange(1,13).reshape(2,6)
print(ar)
```

# Ravel 

Ravel function convert 2D Array in to 1D
